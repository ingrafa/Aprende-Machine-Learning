```python
# Dibuje múltiples gráficos usando bucles for usando sintaxis orientada a objetos
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import seed, randint
seed(100)

# Crear figura y subFiguras
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=True, dpi=120)

# Definir los colores y marcadores a utilizar
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Trazar cada eje
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('Cuatro subtramas en una figura', verticalalignment='bottom', fontsize=10)    
plt.tight_layout()
plt.show()
```
